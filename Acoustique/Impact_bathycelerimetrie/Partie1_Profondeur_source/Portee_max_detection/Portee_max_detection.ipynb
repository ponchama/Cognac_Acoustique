```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# sonar equation
SL = 185                 # Souce level (dB)
NL = 82                  # Noise level (dB)
DI = 0                   # Directivity index (dB)
B = 200                  # bandwidth (Hz)
T = 1                    # transmitted signal duration (s)
PG = 10.0*np.log10(B*T)  # Processing gain (dB)
#RT  = SL - TL - NL + DI + PG 

seuil_d = 10.
```

```python
alpha3 = 0.19/1000.    #dB/m 
alpha12 = 1.60/1000.   #dB/m

NL3 = 82.    #dB
NL12 = 71.   #dB
```

```python
R = np.arange(0.1,120.1,0.1)*1000. #m
```

```python
TL3 = 20*np.log10(R) + alpha3*R
TL12 = 20*np.log10(R)+alpha12*R

RL3 = SL - TL3 - NL3 + DI + PG
RL12 = SL - TL12 -NL12 + DI + PG
```

```python
plt.figure(figsize=(8,5))
plt.plot(R/1000., RL3, 'r', label='f = 3kHz')
plt.plot(R/1000., RL12, 'b', label='f = 12kHz')
plt.plot([-5.,125.], [8.,8.], '-.k', label='seuil à 8dB')
plt.title("Niveau reçu en fonction de la distance parcourue", fontsize=16)
plt.xlabel('distance à la source (km)', fontsize=14)
plt.ylabel('RL (dB)', fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlim([-2.,122.])
plt.ylim(ymin=-100.)
plt.legend(fontsize=13)
plt.grid()

plt.savefig('portee_max', dpi=100.)
```
