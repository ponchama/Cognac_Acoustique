### Ce notebook a pour but de déterminer l'incertitude liée à la position du transducteur de la source par rapport à la bouée de surface (partie 3.1.2 du rapport).

Les calculs sont basés sur la formulation présentée dans le rapport.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### variables : 

g = 9.81 # m^2/s
rhow = 1000 # kg/m^3

# input parameters
h = 100. # m
dUdz = .1/h 

# surface mooring
Sb = 1.*.5 # m^2
Cb = .5

# transductor
m = 10 # kg
rhow_Vt = 5 # kg
St = .2*.5 # m^2
Ct = .5

#
def get_dx(h=100., dUdz=.1/100.):
    r = Ct*St/(Cb*Sb)
    Dt = Ct*rhow*St*h**2*dUdz**2/(1+r**2)
    dx = h*Dt/(m-rhow_Vt)/g
    return dx
    
#print(get_dx())
```

```python
## écart de position entre la bouée et le transducteur en fonction de la longueur de câble (h)

m = 10. 
h = np.arange(10.,201.,5.)
plt.figure(figsize=(8,5))
plt.plot(h,get_dx(h=h), linewidth=2.2)
plt.title('Ecart de position en fonction de la longueur de câble', fontsize=16)
plt.xlabel('longueur du câble (m)', fontsize=14)
plt.ylabel('dx (m)', fontsize=14)
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
#plt.savefig('dx_cable2', dpi=100.)
```

```python
## écart de position entre la bouée et le transducteur en fonction du cisaillement (dUdz)
m=10.
dUdz = np.arange(.01,.5,.01)/100.
plt.figure(figsize=(8,5))
plt.plot(dUdz,get_dx(h=100., dUdz=dUdz), linewidth=2.2)
plt.title('Ecart de position en fonction du cisaillement', fontsize=16)
plt.xlabel('cisaillement (s-1)', fontsize=14)
plt.ylabel('dx (m)', fontsize=14)
plt.grid()
```

```python
plt.figure(figsize=(8,5))

m = 8. # kg
dUdz = np.arange(.01,.501,.01)/100.
plt.plot(dUdz,get_dx(h=100., dUdz=dUdz), label = 'm = 8kg', linewidth=2.2)

m = 10. # kg
dUdz = np.arange(.01,.501,.01)/100.
plt.plot(dUdz,get_dx(h=100., dUdz=dUdz), label = 'm = 10kg', linewidth=2.2)

m = 15. # kg
dUdz = np.arange(.01,.501,.01)/100.
plt.plot(dUdz,get_dx(h=100., dUdz=dUdz), label = 'm = 15kg', linewidth=2.2)

m = 20. # kg
dUdz = np.arange(.01,.501,.01)/100.
plt.plot(dUdz,get_dx(h=100., dUdz=dUdz), label = 'm = 20kg', linewidth=2.2)


plt.title('Ecart de position en fonction du cisaillement \n et de la masse du transducteur', fontsize=16)
plt.xlabel('cisaillement (s-1)', fontsize=14)
plt.ylabel('dx (m)', fontsize=14)
plt.grid()
plt.legend(fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
#plt.savefig('dx_cisail_poids', dpi=100.)
```
