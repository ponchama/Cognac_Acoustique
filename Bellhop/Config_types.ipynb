# Profils de Méditerranée

```python
s = bellhop({'med': 'SSP_4profils.mat'},name='med')
```

```python
c = s.SSP['med']['c']
d = s.SSP['med']['depth']
```

```python
plt.figure(figsize=(8,5))
plt.plot(c[0,:], d, label='Profil n°1 : thermocline à 60m')
plt.plot(c[1,:], d, label='Profil n°2 : profil hiver')
plt.plot(c[2,:], d, label='Profil n°3 : thermocline à 20m')
plt.plot(c[3,:], d, label="Profil n°4 : profil fin d'été")

plt.title ('Profils de célérité de Méditerranée', fontsize=16)
plt.xlabel('célérité (m/s)', fontsize=14)
plt.ylabel('profondeur (m)', fontsize=14 )
plt.gca().invert_yaxis()
plt.grid()
plt.legend(fontsize=12)

plt.savefig('profils_med', dpi=100)
```
